```python
#QQQQQQQQQQQQQ what is the most sold product ?
products = df['Product Name'].value_counts()
most_sold_product=products.sort_values(ascending=False).iloc[[0]]
most_sold_product
```

```python
#QQQQQQQQQQ who is the customer who made the most purchases?
customers = df['Customer Name'].value_counts()
top_customer = customers.sort_values(ascending=False).iloc[[0]]
top_customer
```

```python
#qqqqqqqqqqqqq What is the total sales for each sub-category?
subcategory_sales = df.groupby('Sub-Category')['Sales'].sum()
subcategory_sales
```

```python
#QQQQQQQQWhat is the total sales for each customer?

customer_sales = df.groupby('Customer Name')['Sales'].sum()
customer_sales
```

```python
#QQQQQQQQQQQwhat is the overall sales for each product 
product_sales = df.groupby('Product Name')['Sales'].sum()
product_sales
```

```python

#QQQQQQQQQQQQQQQWhat is the total sales for each region?

Region_sales = df.groupby('Region')['Sales'].sum()
Region_sales
```

```python
#QQQQQQQQQQQWhat is State that has most purchases?

state = df['State'].value_counts()
top_state = state.sort_values(ascending=False).iloc[[0]]
top_state
```

```python
#QQQQQQQQQQQWhat is the average sales per category?
category_avg_sales = df.groupby('Category')['Sales'].mean()
category_avg_sales
```

```python
#QQQQQQQQQWhat is the total sales for each product category and sub-category ?

product_sales = df.groupby(['Category', 'Sub-Category'])['Sales'].sum()
print(product_sales)
```

```python
#QQQQQQQQWhat is the total sales for the 'Technology' category in the 'West' region?
tech_sales_west = df[(df['Category'] == 'Technology') & (df['Region'] == 'West')]['Sales'].sum()
tech_sales_west
```

```python
#QQQQQQQQQQQHow many orders were placed by each customer segment in the state of California?
california_orders = df[df['State'] == 'California']
orders_per_segment_california = california_orders['Segment'].value_counts()
orders_per_segment_california
```

```python
#QQQQQQQQHow many orders were placed by each customer in the Furniture category?
furniture_orders = df[df['Category'] == 'Furniture']['Customer Name'].value_counts()
furniture_orders
```

```python
#QQQQQQQQQQQQQQQ Mention the customer name who bought chairs in the city of Los Angeles
customer_chairs = df[(df['Sub-Category'] == 'Chairs') & (df['City'] == 'Los Angeles')]['Customer Name'].value_counts()
customer_chairs
```

```python
#QQQQQQQQQQQHow many orders were placed by each customer segment in the state of California?
california_orders = df[df['State'] == 'California']['Segment'].value_counts()
california_orders
```
